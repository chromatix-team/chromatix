```python
import chromatix.functional as cf
import matplotlib.pyplot as plt
from functools import partial

import jax.numpy as jnp
```

```python
field = cf.plane_wave(
    (512, 512),
    1.0,
    0.532,
    1.0,
    amplitude=cf.right_circular(),
    power=1.0,
    pupil=partial(cf.square_pupil, w=400.0),
    scalar=False,
)
```

```python
plt.imshow(field.intensity.squeeze())
plt.colorbar()
```

```python
field_after = cf.right_circular_polarizer(field)
```

```python
jnp.allclose(field_after.u, field.u)
```

```python
field_after = cf.left_circular_polarizer(field)
```

```python
plt.imshow(field_after.intensity.squeeze())
plt.colorbar()
```
